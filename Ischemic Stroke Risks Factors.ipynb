# Description of the problem

**Introduction**:

This research aims to determine the association between certain biochemical indicators and the incidence of ischemic stroke. It is an investigation focused on ischemic stroke within the field of neurology.

**Data**:

The data were obtained from the Harvard Dataverse, where all relevant information was exclusively derived from the medical records of Fuzhou People's Hospital in China. The case group comprised 130 patients admitted to the Neurology Department of Fuzhou People's Hospital for the treatment of ischemic stroke.
Source: https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/U1ELA6

Here is the study conducted by the authors who published the data: https://jnsbm.org/article-view/?id=226
XU Y, Hwa W E, Nordin R B, Mohaideen A
K M, Earnest B S P, How W Y, Le Huahui L. Risk Factors and Gender
Difference in the Prediction of Ischemic Stroke in China. J Nat Sc Biol
Med 2023;14:141-150.

**Approach**:

In this project, a binary classification study will be conducted to determine whether the patient is at risk of experiencing an ischemic stroke and to understand the most important risk factors. Standard classification algorithms, ensemble algorithms, and Boosting algorithms will be utilized for this purpose.



# Column description and analysis of the used values:

In this project, an analysis was conducted on the variables and their units of measurement to understand cases of outliers or misentries. Many of these variables were not described in the original paper where they were used and were obtained from other sources.


- **Disease**  0/1 = non-ischemic stroke/ischemic stroke

- **Gender** 0/1 = female/male

- **UA** = Uric Acid - The units of measurement were not found in the paper, but it seems to be µmol/L based on the value range. Normal values typically range from 3.5 to 7.2 (mg/dL), equivalent to 208.2 to 428.3 µmol/L according to [ucsfhealth.org](https://www.ucsfhealth.org/medical-tests/uric-acid----blood-#:~:text=Normal%20values%20range%20between%203.5,for%20results%20for%20these%20tests)

- **Hcy** = Homocysteine - Units of measurement are μmol/L. Normal homocysteine values range from 15 to 100 μmol/L. [urmc.rochester.edu](https://www.urmc.rochester.edu/encyclopedia/content.aspx?ContentTypeID=167&ContentID=homocysteine)

- **Scr** = Serum Creatinine - Units of measurement are μmol/L. Normal creatinine values range from 61.9 to 114.9 µmol/L in men and 53 to 97.2 µmol/L in women. [mountsinai.org](https://www.mountsinai.org/health-library/tests/creatinine-blood-test#:~:text=Normal%20Results,person's%20size%20and%20muscle%20mass)

- **TBIL** = Total Bilirubin -  Units of measurement seem to be μmol/L. Normal total bilirubin values are 1.71 to 20.5 µmol/L. [mountsinai.org](https://www.mountsinai.org/health-library/tests/bilirubin-blood-test)

- **MPV** = Mean Platelet Volume - Commonly measured in femtoliters (fL), with a normal range of 7.4 to 10 fL. 9 fL. [revespcardiol.org](https://www.revespcardiol.org/es-influencia-del-volumen-plaquetario-medio-articulo-X0300893298003593#:~:text=Los%20valores%20medios%20del%20VPM,%2C4%2D10%20fl)

- **P-LCR** = Platelet Large Cell Ratio - Expressed as a percentage, with normal values usually ranging from 15% to 35%. [ncbi.nlm.nih.gov](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7710198/#:~:text=PLCR%20tells%20about%20the%20percentage%20of%20the%20larger%20platelets%20in%20circulation.&text=Normal%20PLCR%20in%20healthy%20individual%20ranges%20from%2015%25%20to%2035%25.&text=The%20values%20of%20these%20indices%20are%20same%20in%20males%20and%20females.&text=A%20screening%20test%20is%20a,of%20a%20disease%20in%20them.)

[healthmatters.io](https://healthmatters.io/understand-blood-test-results/rdw-sd-red-cell-distribution-width)

- **RDW-CV** = Coefficient of Variation of Red Blood Cell Volume Distribution Width - Commonly expressed as a percentage, with normal values in adults ranging from 11.6% to 14.6%. [healthmatters.io](https://healthmatters.io/understand-blood-test-results/rdw-sd-red-cell-distribution-width)

- **FBS** = Fasting Blood Sugar - Measured in mmol/L. Normal values typically range from 3.9 mmol/L to 5.6 mmol/L. [who.int](https://www.who.int/data/gho/indicator-metadata-registry/imr-details/2380#:~:text=The%20expected%20values%20for%20normal,and%20monitoring%20glycemia%20are%20recommended.)

- **TC** = Total cholesterol - Based on the loaded values, units seem to be mmol/L. Normal values are commonly below 5 mmol/L.

### Getting better parameters and a better imputer

```python
X_copy,  y_copy = get_x_y_copy()
numerical_cols = [cname for cname in X_copy.columns if X_copy[cname].dtype in ['float64']]
other_cols = [cname for cname in X_copy.columns if X_copy[cname].dtype not in ['float64']]
```

I am going to use two metrics, 'accuracy' and 'main_score,' to see how the parameters vary in each case. I create the variable 'main_score' to experiment with different main metrics. In the final setup, I chose to use the fbeta scorer as the main score because it is more flexible for adjusting the emphasis on recall. In this case, selecting a beta of 2 gives more weight to recall than precision (considering that a beta greater than 1 already starts to prioritize recall).

I mainly use accuracy to get a general sense of the chosen parameters, but the main_score is the one that will matter most when making parameter selections.

```python
from sklearn.metrics import fbeta_score, make_scorer


fbeta_scorer = make_scorer(fbeta_score, beta=2)
main_score = fbeta_scorer
```

A ColumnTransformer is used to apply normalization and imputation only to numerical variables and not to boolean variables.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from numpy import mean, std

def get_best_imputation_params(scoring="accuracy"):
  num_transform = Pipeline([('preprocessing', MinMaxScaler()),('imputer',SimpleImputer())])
  preprocessor = ColumnTransformer(
    transformers = [
        ('num', num_transform, numerical_cols),
        ('others', 'passthrough', other_cols)
  ])
  pipe = Pipeline(steps=[
      ('preprocessor', preprocessor),
      ('model', RandomForestClassifier(random_state=SEED))
  ])

  cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=SEED)
  param_grid = [
      { 'preprocessor__num__preprocessing': [StandardScaler(), MinMaxScaler()],
        'preprocessor__num__imputer': [SimpleImputer()],
        'model__random_state':[0]
      },
      {
        'preprocessor__num__preprocessing': [MinMaxScaler()],
        'preprocessor__num__imputer__n_neighbors':  [ 5, 10, 50, 100, 150, 500],
        'preprocessor__num__imputer__weights':['uniform', 'distance'],
        'preprocessor__num__imputer':[KNNImputer()],
        'model__random_state':[0]
      }
  ]

  grid = GridSearchCV(pipe, param_grid, cv=cv, scoring=scoring, n_jobs=1)
  grid.fit(X_copy, y_copy)

  print(f"Best params using {scoring} :\n{grid.best_params_}\n")
  print("Best cross-validation score: {:.2f}".format(grid.best_score_))
  return grid
```

### Getting best parameters using scoring: accuracy.

```python
r_accuracy = get_best_imputation_params()
results_df = pd.DataFrame(r_accuracy.cv_results_)
results_df = results_df.sort_values(by=["rank_test_score"])
results_df.head()
```

# Hyperparameters tunning

The hyperparameter tuning is performed using 2 scores to understand how each algorithm varies when choosing between accuracy and main_score: fbeta. Accuracy is used as a general measure to observe how each technique differs based on the score and gain a better understanding of their functioning. However, fbeta will be taken as the main metric for comparison.

Adding the chosen method for data preprocessing.

```python
def create_preprocessor(numerical_cols=numerical_cols, other_cols=other_cols):
  num_transform = Pipeline([('preprocessing', StandardScaler()),('imputer',SimpleImputer())])
  preprocessor = ColumnTransformer(
      transformers = [
          ('num', num_transform, numerical_cols),
          ('others', 'passthrough', other_cols)
  ])
  return preprocessor

```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

def lr_hyperparameter_tunning(scoring="accuracy"):
  lr_model = LogisticRegression(max_iter=1000, random_state=SEED)
  pipe = Pipeline(steps=[('preprocessor', create_preprocessor()),
                         ('lr', lr_model)
         ])
  cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=SEED)

  param_grid = [
  {
      'lr__solver':['liblinear', ],
      'lr__penalty':['l1','l2']
  },
  {
       'lr__solver':[ 'newton-cg', 'newton-cholesky','lbfgs'],
      'lr__penalty':['l2']
  }
  ]

  grid = GridSearchCV(pipe, param_grid, cv=cv, scoring=scoring, n_jobs=1)
  grid.fit(X_copy, y_copy)

  print("Best params:\n{}\n".format(grid.best_params_))
  print(f"Better cross-validation {scoring} score: {round(grid.best_score_,2)}")
```

```python
lr_hyperparameter_tunning()
```

## Adding method for using Cross-Validation, preprocessing and imputer selected

```python
def test_models_cv(model, X,y, scoring="accuracy"):
  pipe = Pipeline(steps=[('preprocessor', create_preprocessor()),
                         ('model', model)
         ])

  cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=SEED)
  scores = cross_val_score(pipe, X, y, scoring=scoring, cv=cv, n_jobs=-1)
  print(f"{scoring} average: {round(mean(scores)*100, 3)}% ,std: {round(std(scores)*100, 3)}%")
  return scores
```

## I search for better features using different techniques

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif
from sklearn.feature_selection import RFECV

def fs_kbest(X,y, scoring="accuracy"):
  pipe = Pipeline(steps=[('preprocessor', create_preprocessor()),
                         ('fs', "passthrough"),
                         ("classify", LinearSVC(dual=False, max_iter=10000)),
         ])

  cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=SEED)
  parameters = [
      {
          'fs': [SelectKBest()],
          'fs__k':[5,10,15,18,19,20],
          'fs__score_func':[mutual_info_classif,f_classif]
      },
      {
          'fs': [RFECV(estimator =DecisionTreeClassifier(), scoring=scoring, cv=cv, n_jobs=-1)],
          'fs__estimator':[DecisionTreeClassifier(criterion="entropy",random_state=SEED),DecisionTreeClassifier(criterion="gini",random_state=SEED)],
      }
  ]
  grid_search = GridSearchCV(pipe, scoring=scoring, param_grid=parameters, cv=cv, n_jobs=1)
  grid_search.fit(X, y)

  print(f"Best score with {scoring}: {round(grid_search.best_score_ *100,2)}")
  print("Better parameters:")
  print(grid_search.best_params_)
  return grid_search
```

# Comparing different models with training and testing data

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, auc
from matplotlib import pyplot

def plot_confusion_matrix(predicciones, resultados):
  cf_matrix = metrics.confusion_matrix(resultados, predicciones)
  group_names = ['Low Risk','False positive','False negative','High Risk']
  plt.figure(figsize=(4, 3))
  group_percentages = ["{0:.2%}".format(value) for value in
                      cf_matrix.flatten()/np.sum(cf_matrix)]

  labels = [f"{v1}\n{v2}" for v1, v2 in
            zip(group_names,group_percentages)]

  labels = np.asarray(labels).reshape(2,2)

  ax = sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues', cbar=False)

  ax.set_title('Confusion Matrix\n',fontsize=10, y=1.1);
  ax.set_xlabel('\nPredicted values"',fontsize=8)
  ax.set_ylabel('Actual values',fontsize=8);
  ax.xaxis.set_ticklabels(['False','True'],fontsize=8)
  ax.yaxis.set_ticklabels(['False','True'],fontsize=8)
  plt.tight_layout()
  plt.show()

def plot_roc_auc(y_test, y_pred):
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)

    # Create the ROC curve
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')

    # Plot the diagonal line (AUC = 0.5)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='AUC = 0.5')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()


def evaluate_model(model, X, y, roc=True, verbose=True):
    numerical = [cname for cname in X.columns if X[cname].dtype in ['float64']]
    other = [cname for cname in X.columns if X[cname].dtype not in ['float64']]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=SEED)
    pipeline = Pipeline([
        ('preprocessor', create_preprocessor(numerical,other)),
        ('model', model)
    ])

    pipeline.fit(X_train, y_train)
    if isinstance(pipeline.named_steps['model'], xgb.XGBClassifier):
      xgb.plot_importance(pipeline.named_steps['model'])
      pyplot.show()

    y_pred = pipeline.predict(X_test)
    #Metric
    accuracy = round(accuracy_score(y_test, y_pred),3)
    precision = round(precision_score(y_test, y_pred),3)
    recall = round(recall_score(y_test, y_pred),3)
    f1 = round(f1_score(y_test, y_pred),3)
    fbeta = round(fbeta_score(y_test, y_pred, beta=2),3)
    results = [accuracy,precision, recall, f1, fbeta]

    if verbose:
      print(f'accuracy: {accuracy}')
      print(f'precision: {precision}')
      print(f'recall: {recall}')
      print(f'f1: {f1}')
      print(f'fbeta: {fbeta}')

      plot_confusion_matrix(y_pred, y_test)

    if roc:
      print("\n")
      plot_roc_auc(y_test,  y_pred)

    return results
```
